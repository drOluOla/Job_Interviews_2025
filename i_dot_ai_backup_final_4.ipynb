### **Voice Agent Pipeline Design**

Below is a brief summary of the intelligent routing and text/voice interaction process using multiple AI agents for a simulated work place scenario.

-  User Input (Speech): The user speaks into the system (input audio).

- Speech to Text: The spoken input is transcribed into text.

- HR Manager Agent (Triage): Analyzes the transcribed text and routes it to the appropriate specialized agent based on intent:

  - Line Manager Agent: For task and role-specific queries using the Get Next Task tool.

  - Colleague Agent: For general workplace tips using the RAG (Retrieval-Augmented Generation) tool.

  - IT Staff Agent: For troubleshooting and tech-related queries using the Computer Use tool.

- Response Generation: The selected agent processes the query and generates a text response.

- Text to Speech: The response is converted back into audio.

- LLM Output (Speech): The synthesized audio is delivered to the user.

```python
# ==============================================================================
# LOGGING FUNCTION FOR KEY PARAMS
# ==============================================================================

def log_conversation_event(conversation_id: str, log_type: str, data: dict):
    """Log conversation events to in-memory dictionary"""
    if conversation_id not in CONVERSATION_LOGS:
        CONVERSATION_LOGS[conversation_id] = {
            "conversation_id": conversation_id,
            "start_time": datetime.now().isoformat(),
            "events": []
        }

    log_entry = {
        "timestamp": datetime.now().isoformat(),
        "type": log_type,
        "data": data
    }

    CONVERSATION_LOGS[conversation_id]["events"].append(log_entry)

# ==============================================================================
# SPEECH GENERATION
# ==============================================================================

def generate_speech(text: str, agent_name: str = "HR Manager") -> bytes:
    """
    Generate speech audio using OpenAI TTS with agent-specific voice characteristics.

    Args:
        text: Text content to convert to speech
        agent_name: Name of the agent (determines voice characteristics)

    Returns:
        Audio content as bytes, or None if generation fails
    """
    try:
        # Get voice configuration for the specified agent
        voice_config = AGENT_VOICES.get(agent_name, AGENT_VOICES["HR Manager"])
        print(f"[debug] Generating speech for {agent_name} with voice '{voice_config['voice']}'")

        # Generate speech using OpenAI TTS
        response = openai.audio.speech.create(
            model="gpt-4o-mini-tts",
            voice=voice_config["voice"],
            input=text,
            speed=voice_config["speed"]
        )
        return response.content

    except Exception as e:
        print(f"[ERROR] Speech generation failed: {e}")
        return None


# ==============================================================================
# AVATAR CREATION AND VISUAL REPRESENTATION
# ==============================================================================

def create_agent_avatar(agent_name: str, is_currently_speaking: bool = False) -> str:
    """
    Create HTML avatar representation for any agent with visual speaking indicators.

    Args:
        agent_name: Name of the agent to create avatar for
        is_currently_speaking: Whether this agent is currently active/speaking

    Returns:
        HTML string containing the styled avatar
    """
    # Get agent configuration (fallback to HR Manager if not found)
    agent_config = AGENT_VOICES.get(agent_name, AGENT_VOICES["HR Manager"])

    # Apply different styling based on whether agent is currently speaking
    if is_currently_speaking:
        # Active agent gets glowing border and pulse animation
        border_styling = f"box-shadow: 0 0 20px {agent_config['color']}; border-color: {agent_config['color']}; animation: pulse 1s infinite;"
        background_overlay = f"{agent_config['color']}20"
        speaking_indicator = '<div style="margin-top: 10px; color: #4CAF50; font-weight: bold;"> SPEAKING</div>'
    else:
        # Inactive agents have standard styling
        border_styling = ""
        background_overlay = "rgba(255, 255, 255, 0.05)"
        speaking_indicator = ""

    # Return complete HTML for agent avatar
    return f"""<style>
                  @keyframes pulse {{
                  0% {{ opacity: 1; }}
                  50% {{ opacity: 0.7; }}
                  100% {{ opacity: 1; }}
                  }}
               </style>
               <div style="
                  text-align: center;
                  padding: 20px;
                  border-radius: 15px;
                  background: linear-gradient(135deg, {agent_config['color']}20, {agent_config['color']}10);
                  border: 3px solid {agent_config['color']};
                  {border_styling}
                  transition: all 0.3s ease;
                  height: 220px;
                  display: flex;
                  flex-direction: column;
                  justify-content: center;
                  align-items: center;
                  position: relative;
                  overflow: hidden;
                ">
                <div style="
                      position: absolute;
                      top: 0;
                      left: 0;
                      right: 0;
                      bottom: 0;
                      background: {background_overlay};
                      z-index: 1;
                "></div>
                <div style="position: relative; z-index: 2;">
                      <div style="font-size: 4em; margin-bottom: 10px;">{agent_config['emoji']}</div>
                      <div style="font-weight: bold; font-size: 1.2em; color: #ffffff; margin-bottom: 5px;">
                          {agent_name}
                      </div>
                      <div style="color: #666; font-size: 0.9em; margin-bottom: 5px;">
                          {agent_config['description']}
                      </div>
                      <div style="color: #888; font-size: 0.8em;">
                          {agent_config['specialty']}
                      </div>
                      {speaking_indicator}
                </div>
              </div>
            """

# ==============================================================================
# CONVERSATION STATE MANAGEMENT
# ==============================================================================
@dataclass
class ConversationData:
    """
    Manages shared state and conversation history across all agents in a conversation.
    Tracks which agent is active, conversation flow, and interaction metadata.
    """
    # Current state
    active_agent_type: Optional[str] = None
    conversation_history: list = field(default_factory=list)
    previous_agent_type: Optional[str] = None

    # Interaction tracking
    last_transcript: str = ""
    turn_count: int = 0
    last_messages: Dict[str, str] = field(default_factory=dict)

    def log_response(self, agent_name: str, message: str):
        """
        Record an agent's response and update conversation state.

        Args:
            agent_name: Name of the agent providing the response
            message: The response message content
        """
        self.active_agent_type = agent_name
        self.conversation_history.append({"agent": agent_name, "message": message})
        self.last_messages[agent_name] = message
        self.turn_count += 1


class ConversationTracker:
    """
    Manages conversation turns and provides interface to conversation data.
    Each turn maintains its own conversation state and agent interactions.
    """

    def __init__(self):
        """Initialize a new conversation with fresh data."""
        self.conversation_data = ConversationData()

    def log_response(self, agent_name: str, message: str):
        """Log an agent response to this conversation."""
        self.conversation_data.log_response(agent_name, message)

    def get_conversation_data(self) -> ConversationData:
        """Get the current data for this conversation."""
        return self.conversation_data


# ==============================================================================
# AUDIO PROCESSING UTILITIES
# ==============================================================================

def preprocess_audio(raw_audio_data, original_sample_rate):
    """
    Clean and normalize audio data for optimal transcription quality.

    Handles format conversion, normalization, resampling, and mono conversion.

    Args:
        raw_audio_data: Raw audio data in various formats
        original_sample_rate: Original sample rate of the audio

    Returns:
        Tuple of (processed_audio_array, target_sample_rate)
    """
    # Ensure we have a numpy array
    if not isinstance(raw_audio_data, np.ndarray):
        raw_audio_data = np.array(raw_audio_data)

    # Convert different integer formats to float32
    if raw_audio_data.dtype == np.int16:
        raw_audio_data = raw_audio_data.astype(np.float32) / 32768.0
    elif raw_audio_data.dtype == np.int32:
        raw_audio_data = raw_audio_data.astype(np.float32) / 2147483648.0
    elif raw_audio_data.dtype != np.float32:
        raw_audio_data = raw_audio_data.astype(np.float32)

    # Convert stereo to mono if needed
    if len(raw_audio_data.shape) > 1:
        raw_audio_data = np.mean(raw_audio_data, axis=1)

    # Normalize audio level (prevent clipping, maintain quality)
    max_amplitude = np.max(np.abs(raw_audio_data))
    if max_amplitude > 0:
        raw_audio_data = raw_audio_data / max_amplitude * 0.95

    # Resample to 16kHz if needed (Whisper's preferred rate)
    target_sample_rate = 16000
    if original_sample_rate != target_sample_rate:
        target_length = int(len(raw_audio_data) * target_sample_rate / original_sample_rate)
        raw_audio_data = np.interp(
            np.linspace(0, len(raw_audio_data), target_length),
            np.arange(len(raw_audio_data)),
            raw_audio_data
        )

    return raw_audio_data, target_sample_rate


def transcribe_audio_openai(processed_audio_data, sample_rate=16000):
    """
    Transcribe audio to text using OpenAI's Whisper model.

    Args:
        processed_audio_data: Clean audio data as numpy array
        sample_rate: Sample rate of the audio (default 16000)

    Returns:
        Transcribed text string, or None if transcription fails
    """
    try:
        # Convert float audio back to 16-bit integer for WAV format
        audio_as_int16 = (processed_audio_data * 32767).astype(np.int16)
        wav_buffer = io.BytesIO()

        # Create WAV file in memory
        with wave.open(wav_buffer, 'wb') as wav_file:
            wav_file.setnchannels(1)  # Mono
            wav_file.setsampwidth(2)  # 16-bit
            wav_file.setframerate(sample_rate)
            wav_file.writeframes(audio_as_int16.tobytes())

        # Prepare buffer for OpenAI API
        wav_buffer.seek(0)
        wav_buffer.name = "audio.wav"  # Required by OpenAI API

        # Call OpenAI Whisper for transcription
        response = openai.audio.transcriptions.create(
            model="whisper-1",
            file=wav_buffer,
            language="en",
            response_format="text",
            temperature=0.2  # Lower temperature for more consistent results
        )

        # Clean up the response
        transcription = response.strip() if isinstance(response, str) else str(response).strip()
        # Transcription (using Whisper) of what user said
        print(f"[debug] User: {transcription}")
        return transcription

    except Exception as e:
        print(f"[ERROR] OpenAI transcription failed: {e}")
        return None


# ==============================================================================
# CORE AGENT INTERACTION LOGIC
# ==============================================================================

# async def get_agent_response(user_input_text: str, conversation_id: str = "default") -> Tuple[str, str]:
#     """
#     Process user input through the multi-agent system starting with HR Manager.

#     The HR Manager acts as the main entry point and may hand off to other agents
#     based on the nature of the user's request.

#     Args:
#         user_input_text: The user's text input
#         conversation_id: Unique identifier for this conversation turn

#     Returns:
#         Tuple of (response_text, active_agent_name)
#     """
#     try:
#         # Get or create conversation tracker
#         if conversation_id not in CONVERSATION_SESSION:
#             CONVERSATION_SESSION[conversation_id] = ConversationTracker()

#         conversation_tracker = CONVERSATION_SESSION[conversation_id]
#         conversation_data = conversation_tracker.get_conversation_data()

#         # Always start with HR Manager as the main entry point
#         # The Runner will handle any necessary handoffs to other agents
#         agent_result = await Runner.run(hr_manager, input=user_input_text)

#         # Determine which agent ultimately responded
#         responding_agent = agent_result.last_agent.name if agent_result.last_agent else "HR Manager"
#         response_text = agent_result.final_output

#         print(f"[debug] Final agent that responded: {responding_agent}")
#         response_text_cleaned = response_text.strip().replace('\n', ' ').replace('\r', '')
#         print(f"[debug] {responding_agent}: {response_text_cleaned}")

#         # Log handoff information for debugging
#         if responding_agent != "HR Manager":
#             print(f"[debug] SUCCESS: Handoff occurred from HR Manager to {responding_agent}")
#         else:
#             print(f"[debug] NO HANDOFF: HR Manager handled the request directly")

#         # Update conversation state
#         conversation_tracker.log_response(responding_agent, response_text)

#         return response_text, responding_agent

#     except Exception as e:
#         print(f"[ERROR] Response generation failed: {e}")
#         import traceback
#         traceback.print_exc()

#         # Return error response while maintaining expected format
#         error_message = f"I apologize, but I'm having trouble processing your request right now. Error: {str(e)}"
#         return error_message, "HR Manager"

async def get_agent_response(user_input_text: str, conversation_id: str = "default") -> Tuple[str, str]:
    """
    Process user input through the multi-agent system starting with HR Manager.
    """
    try:
        # Log user input
        log_conversation_event(conversation_id, "user_input", {
            "text": user_input_text,
            "entry_point": "HR Manager"
        })

        # Get or create conversation tracker
        if conversation_id not in CONVERSATION_SESSION:
            CONVERSATION_SESSION[conversation_id] = ConversationTracker()

        conversation_tracker = CONVERSATION_SESSION[conversation_id]
        conversation_data = conversation_tracker.get_conversation_data()

        # Always start with HR Manager as the main entry point
        agent_result = await Runner.run(hr_manager, input=user_input_text)

        # Determine which agent ultimately responded
        responding_agent = agent_result.last_agent.name if agent_result.last_agent else "HR Manager"
        response_text = agent_result.final_output

        # Log agent response
        log_conversation_event(conversation_id, "agent_response", {
            "responding_agent": responding_agent,
            "response_text": response_text,
            "handoff_occurred": responding_agent != "HR Manager"
        })

        # Log handoff information
        if responding_agent != "HR Manager":
            log_conversation_event(conversation_id, "handoff", {
                "from_agent": "HR Manager",
                "to_agent": responding_agent,
                "success": True
            })
        else:
            log_conversation_event(conversation_id, "no_handoff", {
                "agent": "HR Manager",
                "handled_directly": True
            })

        # Update conversation state
        conversation_tracker.log_response(responding_agent, response_text)

        # Keep debug prints for immediate feedback
        print(f"[debug] Final agent that responded: {responding_agent}")
        response_text_cleaned = response_text.strip().replace('\n', ' ').replace('\r', '')
        print(f"[debug] {responding_agent}: {response_text_cleaned}")

        return response_text, responding_agent

    except Exception as e:
        # Log error
        log_conversation_event(conversation_id, "error", {
            "error_message": str(e),
            "function": "get_agent_response"
        })

        print(f"[ERROR] Response generation failed: {e}")
        import traceback
        traceback.print_exc()

        error_message = f"I apologize, but I'm having trouble processing your request right now. Error: {str(e)}"
        return error_message, "HR Manager"


# ==============================================================================
# VOICE INPUT PROCESSING
# ==============================================================================

def process_voice_input(audio_data, conversation_id: str = "default"):
    """
    Complete pipeline for processing voice input through the onboarding system.

    Steps:
    1. Validate and preprocess audio
    2. Transcribe speech to text
    3. Get agent response (starting with HR Manager)
    4. Generate speech output
    5. Update UI with current agent states

    Args:
        audio_data: Raw audio input (various formats supported)
        conversation_id: identifier for conversation tracking

    Returns:
        Tuple of (output_audio, transcription, status_message, hr_avatar,
                 ai_colleague_avatar, it_avatar, manager_avatar)
    """
    print(f"[debug] Processing voice input with HR Manager as main entry point")

    # Create default avatar states for error cases
    default_avatars = (
        create_agent_avatar("HR Manager"),
        create_agent_avatar("AI Colleague"),
        create_agent_avatar("IT Staff"),
        create_agent_avatar("Line Manager")
    )

    # Validate audio input
    if audio_data is None:
        return (None, "", " No audio data received") + default_avatars

    try:
        # Handle different audio input formats
        if isinstance(audio_data, tuple):
            sample_rate, raw_audio = audio_data
        elif hasattr(audio_data, 'shape'):
            sample_rate = 24000  # Default sample rate
            raw_audio = audio_data
        else:
            return (None, "", "Invalid audio format") + default_avatars

        # Check for empty audio
        if hasattr(raw_audio, 'size') and raw_audio.size == 0:
            return (None, "", "Empty audio data") + default_avatars

        # Preprocess audio for optimal transcription
        processed_audio, final_sample_rate = preprocess_audio(raw_audio, sample_rate)

        # Transcribe speech to text
        transcription = transcribe_audio_openai(processed_audio, final_sample_rate)
        if not transcription:
            return (None, "", "Transcription failed") + default_avatars

        # Get agent response through HR Manager entry point
        loop = asyncio.new_event_loop()
        asyncio.set_event_loop(loop)

        try:
            response_text, active_agent = loop.run_until_complete(
                get_agent_response(transcription, conversation_id)
            )
        finally:
            loop.close()

        # Generate speech output
        speech_audio_bytes = generate_speech(response_text, active_agent)
        output_audio = None
        if speech_audio_bytes:
            try:
                output_audio = (24000, np.frombuffer(speech_audio_bytes, dtype=np.int16))
            except Exception as e:
                print(f"Speech conversion failed: {e}")

        # Get conversation data for status message
        conversation_tracker = CONVERSATION_SESSION.get(conversation_id, ConversationTracker())
        conversation_data = conversation_tracker.get_conversation_data()

        # Create informative status message
        status_message = f"""
                             Conversation Turn #{conversation_data.turn_count}
                             You said: "{transcription}"
                             {active_agent} responded (HR Manager entry point)
                             Speech generated
                             Record your next question and click send audio!
                          """.strip()

        # Update avatars with current speaker highlighted
        hr_avatar = create_agent_avatar("HR Manager", active_agent == "HR Manager")
        ai_colleague_avatar = create_agent_avatar("AI Colleague", active_agent == "AI Colleague")
        it_avatar = create_agent_avatar("IT Staff", active_agent == "IT Staff")
        manager_avatar = create_agent_avatar("Line Manager", active_agent == "Line Manager")

        return (output_audio, transcription, status_message,
                hr_avatar, ai_colleague_avatar, it_avatar, manager_avatar)

    except Exception as e:
        error_message = f"Error: {str(e)}"
        print(f"[ERROR] {error_message}")
        return (None, "", f" {error_message}") + default_avatars


# ==============================================================================
# TEXT INPUT PROCESSING
# ==============================================================================

def process_text_input(user_text: str, conversation_id: str = "default"):
    """
    Process text input through the onboarding system with HR Manager as entry point.

    Args:
        user_text: Text input from the user
        conversation_id: identifier for conversation tracking

    Returns:
        Tuple of (output_audio, input_text, status_message, hr_avatar,
                 ai_colleague_avatar, it_avatar, manager_avatar)
    """
    # Create default avatar states
    default_avatars = (
        create_agent_avatar("HR Manager"),
        create_agent_avatar("AI Colleague"),
        create_agent_avatar("IT Staff"),
        create_agent_avatar("Line Manager")
    )

    # Validate input
    if not user_text.strip():
        return (None, "", " Please enter a message") + default_avatars

    try:
        # Get agent response starting with HR Manager
        loop = asyncio.new_event_loop()
        asyncio.set_event_loop(loop)

        try:
            response_text, active_agent = loop.run_until_complete(
                get_agent_response(user_text, conversation_id)
            )
        finally:
            loop.close()

        # Generate speech output
        speech_audio_bytes = generate_speech(response_text, active_agent)
        output_audio = None
        if speech_audio_bytes:
            output_audio = (24000, np.frombuffer(speech_audio_bytes, dtype=np.int16))

        # Get conversation data
        conversation_tracker = CONVERSATION_SESSION.get(conversation_id, ConversationTracker())
        conversation_data = conversation_tracker.get_conversation_data()

        # Create status message
        status_message = f"""
                             Conversation Turn #{conversation_data.turn_count}
                             You asked: "{user_text}"
                             {active_agent} responded (HR Manager entry point)
                             Speech generated
                             Record your next question and click send audio!
                          """.strip()

        # Update avatars with current speaker highlighted
        hr_avatar = create_agent_avatar("HR Manager", active_agent == "HR Manager")
        ai_colleague_avatar = create_agent_avatar("AI Colleague", active_agent == "AI Colleague")
        it_avatar = create_agent_avatar("IT Staff", active_agent == "IT Staff")
        manager_avatar = create_agent_avatar("Line Manager", active_agent == "Line Manager")

        return (output_audio, user_text, status_message,
                hr_avatar, ai_colleague_avatar, it_avatar, manager_avatar)

    except Exception as e:
        error_message = f" Error: {str(e)}"
        return (None, user_text, error_message) + default_avatars
```

```python
import json
from datetime import datetime
from typing import List, Dict, Any
from deepeval import evaluate
from deepeval.metrics import (
    AnswerRelevancyMetric,
    FaithfulnessMetric,
    ContextualPrecisionMetric,
    ContextualRecallMetric,
    HallucinationMetric,
    BiasMetric,
    ToxicityMetric
)
from deepeval.test_case import LLMTestCase

def parse_conversation_logs(conversation_logs: Dict) -> List[Dict]:
    """Convert conversation logs to a format suitable for DeepEval analysis"""
    parsed_conversations = []

    for conv_id, conv_data in conversation_logs.items():
        if conv_id == 'default':  # Skip tool call logs
            continue

        events = conv_data['events']
        conversation_pairs = []

        # Extract user input and agent response pairs
        for i in range(len(events)):
            if events[i]['type'] == 'user_input':
                user_message = events[i]['data']['text']

                # Find the corresponding agent response
                agent_response = None
                responding_agent = None
                for j in range(i + 1, len(events)):
                    if events[j]['type'] == 'agent_response':
                        agent_response = events[j]['data']['response_text']
                        responding_agent = events[j]['data']['responding_agent']
                        break

                if agent_response:
                    conversation_pairs.append({
                        'user_input': user_message,
                        'agent_response': agent_response,
                        'responding_agent': responding_agent,
                        'timestamp': events[i]['timestamp'],
                        'conversation_id': conv_id
                    })

        parsed_conversations.extend(conversation_pairs)

    return parsed_conversations

def create_test_cases(conversation_pairs: List[Dict]) -> List[LLMTestCase]:
    """Create DeepEval test cases from conversation pairs"""
    test_cases = []

    for pair in conversation_pairs:
        # Create a test case for each conversation turn
        # For hallucination detection, we'll use the user input as context
        # In a real scenario, you might have actual retrieval context or knowledge base content
        context = [
            f"User is asking: {pair['user_input']}",
            f"Agent role: {pair['responding_agent']}",
            "This is an onboarding conversation for a new employee."
        ]

        test_case = LLMTestCase(
            input=pair['user_input'],
            actual_output=pair['agent_response'],
            context=context,  # Required for HallucinationMetric
            # retrieval_context=context  # Alternative parameter name in some versions
        )
        test_cases.append(test_case)

    return test_cases

def evaluate_conversation_quality(conversation_logs: Dict):
    """Evaluate conversation quality using various DeepEval metrics"""

    # Parse conversation logs
    conversation_pairs = parse_conversation_logs(conversation_logs)
    print(f"Found {len(conversation_pairs)} conversation pairs to evaluate")

    # Create test cases
    test_cases = create_test_cases(conversation_pairs)

    # Define metrics to evaluate
    # Start with metrics that don't require special parameters
    basic_metrics = [
        AnswerRelevancyMetric(threshold=0.7),
        BiasMetric(threshold=0.3),
        ToxicityMetric(threshold=0.3)
    ]

    # Metrics that require context
    context_metrics = [
        HallucinationMetric(threshold=0.3),
        # FaithfulnessMetric(threshold=0.7),  # Uncomment if needed
    ]

    all_metrics = basic_metrics + context_metrics

    # Evaluate each test case
    results = []
    for i, test_case in enumerate(test_cases):
        print(f"Evaluating conversation pair {i + 1}/{len(test_cases)}")

        try:
            # Evaluate the test case
            evaluation_result = evaluate(test_cases=[test_case], metrics=all_metrics)

            # Store results with conversation context
            pair_result = {
                'conversation_id': conversation_pairs[i]['conversation_id'],
                'user_input': conversation_pairs[i]['user_input'],
                'agent_response': conversation_pairs[i]['agent_response'],
                'responding_agent': conversation_pairs[i]['responding_agent'],
                'timestamp': conversation_pairs[i]['timestamp'],
                'metrics': {}
            }

            # Extract metric scores
            for metric in all_metrics:
                metric_name = metric.__class__.__name__
                pair_result['metrics'][metric_name] = {
                    'score': getattr(metric, 'score', None),
                    'success': getattr(metric, 'success', None),
                    'reason': getattr(metric, 'reason', None)
                }

            results.append(pair_result)

        except Exception as e:
            print(f"Error evaluating pair {i + 1}: {str(e)}")
            # Add a result with error information
            pair_result = {
                'conversation_id': conversation_pairs[i]['conversation_id'],
                'user_input': conversation_pairs[i]['user_input'],
                'agent_response': conversation_pairs[i]['agent_response'],
                'responding_agent': conversation_pairs[i]['responding_agent'],
                'timestamp': conversation_pairs[i]['timestamp'],
                'metrics': {},
                'error': str(e)
            }
            results.append(pair_result)

    return results

def analyze_results(results: List[Dict]):
    """Analyze and summarize evaluation results"""
    print("\n=== CONVERSATION EVALUATION SUMMARY ===")

    # Overall statistics
    total_conversations = len(results)
    print(f"Total conversation pairs evaluated: {total_conversations}")

    # Metric summaries
    metric_summaries = {}
    for result in results:
        for metric_name, metric_data in result['metrics'].items():
            if metric_name not in metric_summaries:
                metric_summaries[metric_name] = {
                    'scores': [],
                    'successes': 0,
                    'failures': 0
                }

            if metric_data['score'] is not None:
                metric_summaries[metric_name]['scores'].append(metric_data['score'])

            if metric_data['success']:
                metric_summaries[metric_name]['successes'] += 1
            else:
                metric_summaries[metric_name]['failures'] += 1

    # Print metric summaries
    for metric_name, summary in metric_summaries.items():
        print(f"\n{metric_name}:")
        if summary['scores']:
            avg_score = sum(summary['scores']) / len(summary['scores'])
            print(f"  Average score: {avg_score:.3f}")
            print(f"  Score range: {min(summary['scores']):.3f} - {max(summary['scores']):.3f}")
        print(f"  Success rate: {summary['successes']}/{total_conversations} ({100*summary['successes']/total_conversations:.1f}%)")

    # Agent performance breakdown
    agent_performance = {}
    for result in results:
        agent = result['responding_agent']
        if agent not in agent_performance:
            agent_performance[agent] = {'count': 0, 'metrics': {}}

        agent_performance[agent]['count'] += 1
        for metric_name, metric_data in result['metrics'].items():
            if metric_name not in agent_performance[agent]['metrics']:
                agent_performance[agent]['metrics'][metric_name] = []
            if metric_data['score'] is not None:
                agent_performance[agent]['metrics'][metric_name].append(metric_data['score'])

    print(f"\n=== AGENT PERFORMANCE BREAKDOWN ===")
    for agent, perf in agent_performance.items():
        print(f"\n{agent} ({perf['count']} responses):")
        for metric_name, scores in perf['metrics'].items():
            if scores:
                avg_score = sum(scores) / len(scores)
                print(f"  {metric_name}: {avg_score:.3f} avg")

```

```python
import json
from datetime import datetime
from typing import List, Dict, Any
from deepeval import evaluate
import warnings
from deepeval.metrics import (
    AnswerRelevancyMetric,
    BiasMetric,
    ToxicityMetric,
    # ConversationRelevancyMetric,  # Available in newer versions
)
from deepeval.test_case import LLMTestCase

# Disable DeepEval verbose output and promotional messages
os.environ["DEEPEVAL_TELEMETRY_OPT_OUT"] = "YES"
os.environ["DEEPEVAL_CONFIDENT_AI_BANNER"] = "NO"
warnings.filterwarnings("ignore")

def parse_conversation_logs(conversation_logs: Dict) -> List[Dict]:
    """Convert conversation logs to a format suitable for DeepEval analysis"""
    parsed_conversations = []

    for conv_id, conv_data in conversation_logs.items():
        if conv_id == 'default':  # Skip tool call logs
            continue

        events = conv_data['events']
        conversation_pairs = []

        # Extract user input and agent response pairs
        for i in range(len(events)):
            if events[i]['type'] == 'user_input':
                user_message = events[i]['data']['text']

                # Find the corresponding agent response
                agent_response = None
                responding_agent = None
                for j in range(i + 1, len(events)):
                    if events[j]['type'] == 'agent_response':
                        agent_response = events[j]['data']['response_text']
                        responding_agent = events[j]['data']['responding_agent']
                        break

                if agent_response:
                    conversation_pairs.append({
                        'user_input': user_message,
                        'agent_response': agent_response,
                        'responding_agent': responding_agent,
                        'timestamp': events[i]['timestamp'],
                        'conversation_id': conv_id
                    })

        parsed_conversations.extend(conversation_pairs)

    return parsed_conversations

def evaluate_conversation_quality_simple(conversation_logs: Dict):
    """Evaluate conversation quality using metrics that don't require context"""

    # Parse conversation logs
    conversation_pairs = parse_conversation_logs(conversation_logs)
    print(f"Found {len(conversation_pairs)} conversation pairs to evaluate")

    # Define metrics that don't require context
    metrics = [
        AnswerRelevancyMetric(threshold=0.7),
        BiasMetric(threshold=0.3),
        ToxicityMetric(threshold=0.3)
    ]

    # Evaluate each conversation pair
    results = []
    for i, pair in enumerate(conversation_pairs):
        print(f"Evaluating conversation pair {i + 1}/{len(conversation_pairs)}")

        # Create test case for this pair
        test_case = LLMTestCase(
            input=pair['user_input'],
            actual_output=pair['agent_response']
        )

        try:
            # Evaluate the test case
            evaluation_result = evaluate(test_cases=[test_case], metrics=metrics)

            # Store results with conversation context
            pair_result = {
                'conversation_id': pair['conversation_id'],
                'user_input': pair['user_input'],
                'agent_response': pair['agent_response'],
                'responding_agent': pair['responding_agent'],
                'timestamp': pair['timestamp'],
                'metrics': {},
                'evaluation_success': True
            }

            # Extract metric scores
            for metric in metrics:
                metric_name = metric.__class__.__name__
                pair_result['metrics'][metric_name] = {
                    'score': getattr(metric, 'score', None),
                    'success': getattr(metric, 'success', None),
                    'reason': getattr(metric, 'reason', None),
                    'threshold': getattr(metric, 'threshold', None)
                }

            results.append(pair_result)

        except Exception as e:
            print(f"Error evaluating pair {i + 1}: {str(e)}")
            # Add a result with error information
            pair_result = {
                'conversation_id': pair['conversation_id'],
                'user_input': pair['user_input'],
                'agent_response': pair['agent_response'],
                'responding_agent': pair['responding_agent'],
                'timestamp': pair['timestamp'],
                'metrics': {},
                'evaluation_success': False,
                'error': str(e)
            }
            results.append(pair_result)

    return results

def calculate_conversation_flow_metrics(conversation_pairs: List[Dict]) -> Dict:
    """Calculate additional metrics specific to conversation flow"""

    metrics = {
        'total_turns': len(conversation_pairs),
        'avg_response_length': 0,
        'agent_handoffs': 0,
        'response_times': [],
        'agent_distribution': {}
    }

    total_length = 0
    previous_agent = None

    for pair in conversation_pairs:
        # Response length
        response_length = len(pair['agent_response'].split())
        total_length += response_length

        # Agent handoffs
        current_agent = pair['responding_agent']
        if previous_agent and previous_agent != current_agent:
            metrics['agent_handoffs'] += 1
        previous_agent = current_agent

        # Agent distribution
        if current_agent not in metrics['agent_distribution']:
            metrics['agent_distribution'][current_agent] = 0
        metrics['agent_distribution'][current_agent] += 1

    if conversation_pairs:
        metrics['avg_response_length'] = total_length / len(conversation_pairs)

    return metrics

def analyze_results_comprehensive(results: List[Dict]):
    """Comprehensive analysis of evaluation results"""
    print("\n=== COMPREHENSIVE CONVERSATION EVALUATION SUMMARY ===")

    # Filter successful evaluations
    successful_results = [r for r in results if r.get('evaluation_success', True)]
    failed_results = [r for r in results if not r.get('evaluation_success', True)]

    print(f"Total conversation pairs: {len(results)}")
    print(f"Successfully evaluated: {len(successful_results)}")
    print(f"Failed evaluations: {len(failed_results)}")

    if failed_results:
        print("\nFailed evaluations:")
        for result in failed_results:
            print(f"  - {result.get('error', 'Unknown error')}")

    if not successful_results:
        print("No successful evaluations to analyze.")
        return

    # Metric summaries
    metric_summaries = {}
    for result in successful_results:
        for metric_name, metric_data in result['metrics'].items():
            if metric_name not in metric_summaries:
                metric_summaries[metric_name] = {
                    'scores': [],
                    'successes': 0,
                    'failures': 0,
                    'threshold': metric_data.get('threshold')
                }

            if metric_data['score'] is not None:
                metric_summaries[metric_name]['scores'].append(metric_data['score'])

            if metric_data['success']:
                metric_summaries[metric_name]['successes'] += 1
            else:
                metric_summaries[metric_name]['failures'] += 1

    # Print metric summaries
    print(f"\n=== METRIC PERFORMANCE ===")
    for metric_name, summary in metric_summaries.items():
        print(f"\n{metric_name}:")
        if summary['threshold']:
            print(f"  Threshold: {summary['threshold']}")
        if summary['scores']:
            avg_score = sum(summary['scores']) / len(summary['scores'])
            print(f"  Average score: {avg_score:.3f}")
            print(f"  Score range: {min(summary['scores']):.3f} - {max(summary['scores']):.3f}")
        total_evaluated = summary['successes'] + summary['failures']
        print(f"  Success rate: {summary['successes']}/{total_evaluated} ({100*summary['successes']/total_evaluated:.1f}%)")

    # Agent performance breakdown
    agent_performance = {}
    for result in successful_results:
        agent = result['responding_agent']
        if agent not in agent_performance:
            agent_performance[agent] = {'count': 0, 'metrics': {}}

        agent_performance[agent]['count'] += 1
        for metric_name, metric_data in result['metrics'].items():
            if metric_name not in agent_performance[agent]['metrics']:
                agent_performance[agent]['metrics'][metric_name] = {
                    'scores': [],
                    'successes': 0,
                    'failures': 0
                }
            if metric_data['score'] is not None:
                agent_performance[agent]['metrics'][metric_name]['scores'].append(metric_data['score'])
            if metric_data['success']:
                agent_performance[agent]['metrics'][metric_name]['successes'] += 1
            else:
                agent_performance[agent]['metrics'][metric_name]['failures'] += 1

    print(f"\n=== AGENT PERFORMANCE BREAKDOWN ===")
    for agent, perf in agent_performance.items():
        print(f"\n{agent} ({perf['count']} responses):")
        for metric_name, metric_perf in perf['metrics'].items():
            if metric_perf['scores']:
                avg_score = sum(metric_perf['scores']) / len(metric_perf['scores'])
                success_rate = metric_perf['successes'] / (metric_perf['successes'] + metric_perf['failures'])
                print(f"  {metric_name}: {avg_score:.3f} avg score, {success_rate:.1%} success rate")

    # Conversation flow metrics
    conversation_pairs = []
    for result in successful_results:
        conversation_pairs.append({
            'agent_response': result['agent_response'],
            'responding_agent': result['responding_agent']
        })

    flow_metrics = calculate_conversation_flow_metrics(conversation_pairs)

    print(f"\n=== CONVERSATION FLOW METRICS ===")
    print(f"Total conversation turns: {flow_metrics['total_turns']}")
    print(f"Average response length: {flow_metrics['avg_response_length']:.1f} words")
    print(f"Agent handoffs: {flow_metrics['agent_handoffs']}")
    print(f"Agent distribution:")
    for agent, count in flow_metrics['agent_distribution'].items():
        percentage = (count / flow_metrics['total_turns']) * 100
        print(f"  {agent}: {count} responses ({percentage:.1f}%)")

# Example usage
if __name__ == "__main__":

    # Run evaluation with simple metrics (no context required)
    print("Running evaluation with simple metrics...")
    results = evaluate_conversation_quality_simple(CONVERSATION_LOGS)

    # Analyze results
    analyze_results_comprehensive(results)
```
